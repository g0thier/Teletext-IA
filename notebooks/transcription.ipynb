```python
path = '/Users/gauthier/Documents/Travail/Projets/IFG/008 Communauté 0030-2505/02 Parcours Leadership/04 Les conférences-débats/9. Comment gérer le stress ?/Webinaire Parcours Leadership T9 avec Magali HUBERT.mp4'

txt_path = path + ".txt"
```

```python
import whisper

model = whisper.load_model("turbo")
result = model.transcribe(path)
```

```python
# Écriture dans le fichier txt
with open(txt_path, "w", encoding="utf-8") as f:
    for segment in result['segments']:
        parole = segment["text"]
        f.write(parole + "\n")
```
